```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
# Task 1
asia = df[(df['continent'] == 'Asia') & (df['year'].isin([2002, 2007]))]
asia.loc[:, 'year'] = asia.loc[:, 'year'].astype(str)

# ── Step 1: Plotly Express base chart ─────────────────────────────────────────
fig = px.scatter(
    data_frame=asia,
    x='gdpPercap',
    y='lifeExp',
    color='year',                         
    hover_name='country',
    hover_data={'year': False},
    title='Asian nations gained in wealth and longevity 2000-2007 — most shifted right and up',
    labels={'gdpPercap': 'GDP per Capita (log scale, US$)', 'lifeExp': 'Life Expectancy'},
    color_discrete_map={'2002': '#AAAAAA', '2007': '#2E75B6'},  
    log_x=True,
    height=500
)

# ── Step 2: Graph Objects customisation ───────────────────────────────────────
fig.update_traces(
    marker=dict(size=9, opacity=0.85)
)

fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(title='Year', orientation='h', y=1.08),
    margin=dict(l=60, r=40, t=70, b=40),
)

fig.update_xaxes(gridcolor='#EEEEEE')
fig.update_yaxes(gridcolor='#EEEEEE')

fig.show()
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2
# YOUR CODE HERE
import plotly.express as px
import plotly.graph_objects as go

df = px.data.gapminder()

HIGHLIGHT     = 'Asia'
HIGHLIGHT_COLOR = '#E63946'
GREY          = '#DDDDDD'

data_2007 = df[df['year'] == 2007].copy()

# Split into grey (others) and highlighted
others = data_2007[data_2007['continent'] != HIGHLIGHT]
hi     = data_2007[data_2007['continent'] == HIGHLIGHT]

fig = go.Figure()

# ── Grey bubbles first (behind) ───────────────────────────────────────────────
fig.add_trace(go.Scatter(
    x=others['gdpPercap'],
    y=others['lifeExp'],
    mode='markers',
    name='Other continents',
    marker=dict(
        size=others['pop'].apply(lambda p: max(5, min(50, (p / 1_000_000) ** 0.5 * 3))),
        color=GREY,
        opacity=0.6,
        line=dict(width=0.4, color='white')
    ),
    customdata=others[['country', 'continent', 'pop']].values,
    hovertemplate=(
        "<b>%{customdata[0]}</b><br>"
        "Continent: %{customdata[1]}<br>"
        "GDP per capita: $%{x:,.0f}<br>"
        "Life expectancy: %{y:.1f} yrs<br>"
        "Population: %{customdata[2]:,.0f}<extra></extra>"
    )
))

# ── Highlighted continent on top ──────────────────────────────────────────────
fig.add_trace(go.Scatter(
    x=hi['gdpPercap'],
    y=hi['lifeExp'],
    mode='markers',
    name=HIGHLIGHT,
    marker=dict(
        size=hi['pop'].apply(lambda p: max(5, min(50, (p / 1_000_000) ** 0.5 * 3))),
        color=HIGHLIGHT_COLOR,
        opacity=0.85,
        line=dict(width=0.6, color='white')
    ),
    customdata=hi[['country', 'pop']].values,
    hovertemplate=(
        "<b>%{customdata[0]}</b><br>"
        "GDP per capita: $%{x:,.0f}<br>"
        "Life expectancy: %{y:.1f} yrs<br>"
        "Population: %{customdata[1]:,.0f}<extra></extra>"
    )
))

# ── Annotations ───────────────────────────────────────────────────────────────
# Label China and India (the outlier giants)
for country, note in [('China', 'China — massive<br>population, rising GDP'),
                       ('India', 'India — large population,<br>lower GDP per capita'),
                       ('Japan', 'Japan — highest<br>life expectancy in Asia')]:
    row = hi[hi['country'] == country].iloc[0]
    fig.add_annotation(
        x=float(row['gdpPercap']),
        y=float(row['lifeExp']),
        text=f"<b>{country}</b><br><i>{note}</i>" if note else f"<b>{country}</b>",
        showarrow=True,
        arrowhead=2, arrowcolor=HIGHLIGHT_COLOR, arrowwidth=1.5,
        ax=50, ay=-45,
        font=dict(size=10, color=HIGHLIGHT_COLOR),
        bgcolor='rgba(255,255,255,0.8)',
        borderpad=3
    )

# Group annotation
fig.add_annotation(
    x=0.01, y=0.99,
    xref='paper', yref='paper',
    text=(
        "<b>Asia in 2007</b> spans the widest GDP range<br>"
        "of any continent — from under $1k to over $40k<br>"
        "per capita, yet life expectancy is converging upward."
    ),
    showarrow=False,
    xanchor='left', yanchor='top',
    font=dict(size=11, color='#333333'),
    bgcolor='rgba(255,240,240,0.85)',
    bordercolor=HIGHLIGHT_COLOR,
    borderwidth=1, borderpad=6
)

fig.update_layout(
    title=dict(
        text="<b>Asia spans the widest wealth gap of any continent — 2007</b><br>"
             "<sup>Bubble size = population · x-axis log scale · all other continents in grey</sup>",
        font=dict(size=16, family='Arial'),
        x=0.5, xanchor='center'
    ),
    xaxis=dict(
        title='GDP per Capita (log scale, US$)',
        type='log',
        showgrid=True, gridcolor='#EEEEEE',
        tickprefix='$', tickformat=','
    ),
    yaxis=dict(
        title='Life Expectancy (years)',
        showgrid=True, gridcolor='#EEEEEE'
    ),
    legend=dict(
        orientation='h', y=1.08,
        font=dict(size=12)
    ),
    plot_bgcolor='white', paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    margin=dict(l=60, r=40, t=100, b=50),
    height=580, width=980
)


fig.show()
```
